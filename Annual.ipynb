```python
import pandas as pd
```

```python
df=pd.read_csv('Annual-cause-death-numbers-new.csv')
print(df)
```

```python
df.columns = (
    df.columns
    .str.strip()
    .str.replace(r'[\n\r"]+', '', regex=True)  
    .str.replace(r'\s+', ' ', regex=True)      
    .str.strip()
)

print(df.columns.tolist())

df_long = df.melt(
    id_vars=["Entity", "Code", "Year"], 
    var_name="Cause", 
    value_name="Fatalities"
)

print(df_long)




```
